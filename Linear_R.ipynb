## Outliers Analysis

```python
# iqr= q3-q1
# q1=25th percentile
# q2= 50th percentile
# q3= 75th percentile
# 25th percentile is x : 25th percentile of the value are less than or equal to x and 75th percentile values are greater then or equals to x
```

```python
import seaborn as sns 
```

```python
sns.boxplot(y='degree_p', data=df)
```

```python
q1=df['degree_p'].quantile(0.25)
q2=df['degree_p'].quantile(0.5)
q3=df['degree_p'].quantile(0.75)
```

```python
iqr=q3-q1
```

```python
upper_bound=q3+1.5*iqr
lower_bound=q1-1.5*iqr
# some of all values /number of entries
```

```python
upper_bound
lower_bound
```

```python
df=df[(df['degree_p']<upper_bound) & (df['degree_p']>lower_bound)]
```

```python
sns.boxplot(y='degree_p', data=df)
```

```python
column=['ssc_p','hsc_p', 'etest_p', 'mba_p']
for c in column:
    q1=df[c].quantile(0.25)
    q3=df[c].quantile(0.75)
    iqr=q3-q1
    upper_bound=q3+1.5*iqr
    lower_bound=q1-1.5*iqr
    df=df[(df[c]<upper_bound) & (df[c]>lower_bound)]
```

```python
sns.boxplot(y='hsc_p', data=df)
```

## Standardization

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
# X_train and Y_train  will be used to build the model 
# and after the model build we are going to give our X_test to our model and make prediction --> Y_pred
#     now we are going to compare our Y_test with Y_pred and evaluate our mode;
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(X_train, Y_train)
```

```python
y_pred=lr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(Y_test,y_pred)
```

```python
# r square - it tells how well model fits the data [0,1], 1 means the model fits the data well 
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(Y_test, y_pred)
```
