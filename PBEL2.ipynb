```python
def reverse_string(name):
    return name[::-1]
name='madam'  # palindrome 
print(reverse_string(name))
```

```python
def is_palindrome(text):
    return text == text[::-1]
print(is_palindrome('naman'))
```

```python
# vowels count   #asif
def count_vowels(str):
    count=0
    for i in str.lower():
        if i in 'aeiou':
            count+=1
    return count
print(count_vowels("Education"))
            
```

```python
string = input("Enter the word ")
text = string.lower()
def palin(word):
  pal = word[::-1]  # reverse string
  if pal.lower()==text:
    print("yes it is palindrome")
  else:
    print("it is not palindrome")
palin(string)
```
