## outlier Analysis 

```python
# 25th percentile = Q1
# 50th percentile =Q2
# 75th percentile =Q3
# 25th percentile is 25th percentage of the value are less than or equal to x and 75th percetile vlauesare greater than or equa to x
```

```python
import seaborn as sns
```

```python
sns.boxplot(y='degree_p', data=df)
```

```python
q1 = df['degree_p'].quantile(0.25)
q2 = df['degree_p'].quantile(0.50)
q3 = df['degree_p'].quantile(0.75)
```

```python
q3
```

```python
iqr = q3-q1
iqr
```

```python
upper_bound = q3+1.5*iqr
lower_bound = q1-1.5*iqr
```

```python
upper_bound
```

```python
df = df[(df['degree_p']<upper_bound) & (df['degree_p']>lower_bound)]
```

```python
sns.boxplot(y='degree_p', data=df)
```

```python
column=['ssc_p', 'hsc_p', 'etest_p', 'mba_p']
for c in column:
    q1=df[c].quantile(0.25)
    q3=df[c].quantile(0.75)
    iqr=q3-q1
    upper_bound=q3+1.5*iqr
    lower_bound=q1-1.5*iqr
    df=df[(df[c]<upper_bound) & (df[c]>lower_bound)]
```

```python
sns.boxplot(y='hsc_p', data=df)
```
